# Transient to Bronze - Estabelecimento

Os dados de Staging apenas são copiados para Raw para ficarem dentro do nosso ambiente analítico do data lakehouse.

```python
spark.sql("USE CATALOG saude_sus")
```

```python
df_staging = spark.read.table('saude_sus.staging.estabelecimentos')
display(df_staging)
```

```python
df_staging.write \
    .format('delta') \
    .mode('append') \
    .option('mergeSchema', 'true') \
    .saveAsTable('saude_sus.raw.RAW_ESTABELECIMENTO')
```
