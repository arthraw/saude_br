# Transient to Bronze - Tipo Unidade

```python
spark.sql('USE CATALOG saude_sus')
```

```python
df_bronze = spark.read.table('saude_sus.staging.tipo_unidades')
```

```python
df_bronze.write \
    .format('delta') \
    .mode('overwrite') \
    .option('mergeSchema', 'true') \
    .saveAsTable('saude_sus.raw.raw_tipo_unidade')
print('Dados gravados com sucesso na raw: saude_sus.raw.raw_tipo_unidade')
```
